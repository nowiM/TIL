```python
df.isnull().sum()
```

```python
df2 = df.copy()
df2
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df2['label'] = le.fit_transform(df2['label'])

df2
```

```python
df2['label'].unique()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 전처리
X = df[['height', 'weight']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# KMeans 군집화 (3개의 군집으로)
kmeans = KMeans(n_clusters=3, random_state=42)
df2['cluster'] = kmeans.fit_predict(X_scaled)

# 시각화를 위한 실제 라벨과 클러스터 비교
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 실제 label 시각화
sns.scatterplot(data=df2, x='height', y='weight', hue='label', ax=axes[0])
axes[0].set_title('Actual Labels')

# 클러스터 결과 시각화
sns.scatterplot(data=df2, x='height', y='weight', hue='cluster', palette='deep', ax=axes[1])
axes[1].set_title('KMeans Clustering')

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import accuracy_score
from scipy.stats import mode
import numpy as np

# 실제 라벨을 숫자로 변환
label_mapping = {label: idx for idx, label in enumerate(df['label'].unique())}
df['label_encoded'] = df['label'].map(label_mapping)

# 클러스터와 실제 라벨 매핑 (다수 투표 방식)
cluster_to_label = {}
for cluster in range(3):
    most_common = mode(df[df['cluster'] == cluster]['label_encoded'], keepdims=True).mode[0]
    cluster_to_label[cluster] = most_common

# 클러스터 결과를 예측된 라벨로 변환
df['predicted_label'] = df['cluster'].map(cluster_to_label)

# 정확도 계산
accuracy = accuracy_score(df['label_encoded'], df['predicted_label'])

accuracy, cluster_to_label
```

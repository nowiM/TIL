```python
def get_vif(formula, df):
    from patsy import dmatrices
    
    y, X = dmatrices(formula, df, return_type = 'dataframe')
    import pandas as pd
    
    vif = pd.DataFrame()
    
    from statsmodels.stats.outliers_influence import variance_inflation_factor

    vif['VIF Factor'] = [variance_inflation_factor(X.values, i) 
                         for i in range(X.shape[1])]
    vif['features'] = X.columns
    
    return vif 
```

```python
def numerical_derivative(f, x):
    delta_x = 1e-4
    grad = np.zeros_like(x)
    it = np.nditer(x, flags = ['multi_index'], op_flags = ['readwrite'])
    while not it.finished:
        idx = it.multi_index
        tmp_val = x[idx]
        x[idx] = float(tmp_val) + delta_x
        fx1 = f(x) # f(x + delta_x)

        x[idx] = float(tmp_val) - delta_x
        fx2 = f(x)
        grad[idx] = (fx1 - fx2) / (2 * delta_x)
        x[idx] = tmp_val
        it.iternext()

    return grad
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pandas as pd

# 1. 불필요한 열 제거
df = pd.read_csv('../AI데이터/newbie.csv')  # 재로드 (혹시 이전 세션에서 수정되었을 수 있으므로)
df.drop(columns=['who', 'Country', 'Years on Internet'], inplace=True)

# 2. 결측값 처리
# Age: 평균
imp_mean = SimpleImputer(strategy='mean')
df['Age'] = imp_mean.fit_transform(df[['Age']]).ravel()

# 나머지: 최빈값
for col in ['Household Income', 'Sexual Preference', 'Marital Status']:
    imp_freq = SimpleImputer(strategy='most_frequent')
    df[col] = imp_freq.fit_transform(df[[col]]).ravel()

# 3. 범주형 변수 category로 변환 후 더미 처리
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype('category')

df_encoded = pd.get_dummies(df.drop(columns=['Newbie']), drop_first=True)
y = df['Newbie']

# 4. 학습용/평가용 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(df_encoded, y, test_size=0.2, random_state=42)

# 5. 특성 정규화
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 6. 로지스틱 회귀 모델 훈련 및 평가
model = LogisticRegression(max_iter=1000, solver='liblinear')
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)

accuracy
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Household Income']=df['Household Income'].fillna(df['Household Income'].mode()[0])
df['Sexual Preference']=df['Sexual Preference'].fillna(df['Sexual Preference'].mode()[0])
df['Marital Status']=df['Marital Status'].fillna(df['Marital Status'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df = pd.get_dummies(df)
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()
mm_Age = mm.fit_transform(df[["Age"]])
df["Age"] = mm_Age
df.head()
```

```python
x_data=df.iloc[:,1:]
y_data=df['Newbie']
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x_data, y_data, test_size=0.33, random_state=42)
```

```python
model = linear_model.LogisticRegression(solver="liblinear")
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
score = accuracy_score(y_test, pred)
print('test score: ', score)
```

```python
from sklearn.model_selection import GridSearchCV, KFold

estimator = LogisticRegression()
cv = KFold(n_splits=5, random_state=42, shuffle=True)
parameters = {'penalty' : ['l1', 'l2', 'elasticnet', None],
            'solver' : ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
            'max_iter' : [10, 50, 100, 150, 200],
            'random_state' : [7, 23, 42, 78]}
model = GridSearchCV(estimator = estimator,
                    param_grid = parameters,
                    cv = cv, verbose = 1,
                    n_jobs = 1, refit = True)
model.fit(x_train, y_train)
```

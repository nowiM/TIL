```python
feature_cols = ['Press time(ms)', 'Pressure 1', 'Pressure 2', 'Pressure 5']
X = df[feature_cols]

from sklearn.model_selection import train_test_split

# 섞어서 8:2로 분할
X_train, X_test = train_test_split(
    X, test_size=0.2, shuffle=True, random_state=42)

# 스케일링(필수) → 훈련 데이터로 fit 후 두 세트 모두 transform
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)
```

```python
from sklearn.svm import OneClassSVM
model = OneClassSVM(kernel='rbf', gamma=0.1, nu=0.05)
model.fit(X_train)
```

```python
import matplotlib.pyplot as plt
y_pred = model.predict(X_test)          # 1 / -1
scores = model.decision_function(X_test)

print("정상·이상 예측 분포:", np.unique(y_pred, return_counts=True))
plt.hist(scores, bins=50)
plt.title("Decision Function Distribution (test)")
plt.xlabel("score"); plt.ylabel("freq"); plt.show()
```

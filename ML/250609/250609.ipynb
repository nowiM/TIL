# Random Forest

Decision Tree(의사 결정 트리)는 과대 적합이 일어날 가능성이 있=높다. 또한 가지치기를 통해 충분히 해결할 수 없다.
따라서 Random Forest를 통해 보완할 수 있다. 

Random Forest는
새로운 데이터 포인트를 각 트리에
동시에 통과시키며, 각 트리가 분류한 결과에서 투표를 실시하여 가장 많이 득표한 결과를
최종 분류 결과로 선택

```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
iris = datasets.load_iris()
print('아이리스 종류 : ', iris.target_names)
print('target : [0: setosa, 1: versicolor, 2: virginica]')
print('데이터 수 : ', len(iris.data))
print('데이터 열 이름 :', iris.feature_names)
```

```python
data = pd.DataFrame(
    {'sepal length': iris.data[:, 0],
     'sepal width': iris.data[:, 1],
     'petal length': iris.data[:, 2],
     'petal width': iris.data[:, 3],
     'species': iris.target,
    }
)

data.head()
```

```python
data.info()
```

```python
x = data[['sepal length', 'sepal width', 'petal length', 'petal width']]
y = data['species']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state= 42)
```

```python
x_train.shape, y_train.shape
```

```python
forest = RandomForestClassifier(n_estimators = 100) # 100개의 서로 다른 결정 트리 생성
forest.fit(x_train, y_train)
```

```python
y_pred = forest.predict(x_test)
print(y_pred)
print(list(y_test))
```

```python
print('정확도 :', metrics.accuracy_score(y_test, y_pred))
```

```python
# 정확도말고 정밀도, 재현율 등 다양한 것들을 체크해야됨
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

feature_impotances_values = forest.feature_importances_
feature_impotances = pd.Series(feature_impotances_values, index = x_train.columns)
feature_top4 = feature_impotances.sort_values(ascending = False)[:4]
plt.figure(figsize = (8, 6))
plt.title("Top 4 Feature Importances")
sns.barplot(x = feature_top4, y = feature_top4.index)
plt.show()
```

# SVM

SVM을 통해 분류를 진행하였을 때 경계선을 기준으로 잘 분류되면 좋겠지만 그렇지 않은 경우에
**소프트 마진**을 활용한다. SVM에서 소프트 마진을 활용해 경계선을 잘 적하는 것이 핵심이다.

* 인수
    * n_samples : 표본 데이터의 수, 디폴트 100
    * n_features : 독립 변수의 수, 디폴트 2
    * centers : 정수 또는 [n_centers, n_features] 크기의 배열, 생성할 클러스터의 수 혹은 중심, 디폴트 3
    * cluster_std : 클러스터의 펴준 편차, 디폴트 1.0
    * center_box : 생성할 클러스터의 바운스 박스, 디폴트(-10.0, 10.0)
* 반환값
    * X : 독립 변수. [n_samples, n_features] 크기의 배열
    * y : 종속 변수. [n_samples] 크기의 배열
    *

# 두 데이터 간의 경계선 찾기

#### SVC 속성
* n_support_ : 각 클래스의 서포트 벡터의 개수
* support_ : 각 클래스의 서포트 벡터의 인덱스
* support_vectors_ : 각 클래스의 서포트의 x 값
* coef_ : 기울기
* intercept_ : 절편
* dual_coef_ : 각 원소가 ai`yi로 이루어진 벡터. 음의 계수와 양이 계수를 나타낸다.

```python
from sklearn.svm import SVC

# C : 잘못 분류된 데이터로부터 해당 클래스의 결정 경계까지의 거리로 데이터점이 잘못 분류된 
#     정도를 나타낸다.
#     SVM 모델이 오류를 어느정도 허용할 것인지 파라미터 C를 통해 지정
#     C값이 클수록 하다마진(오류 허용 안 함), 작을수록 소프트마진(오류를 허용함)이다.
#     gamma :  결정 경계를 얼마나 유연하게 그을 것인지 정해주는 역할
#     gamma 값을 높이면 학습 데이터에 많이 의존해서 결정 경계를 구불구불 긋게 된다.(오버피팅)
#     gamma 값을 낮추면 학습 데이터에 별로 의존하지 않고 경절 경계를 직선에 가깝게 긋다.(언더피팅)
#     C와 Gamma를 지나치게 높게 잡는다면 오버피팅이 될 수 있다.
#     커널(kernel) : 판별 경계를 찾기위한 내적 함수.
#       'Linear', 'poly', 'rbf', 'sigmoid', 'precomputed', 'default는 rbf'

model = SVC(kernel = 'linear', C = 1e10).fit(X, y)
```

```python
model.n_support_
```

```python
model.support_
```

```python
model.intercept_
```

```python
import numpy as np

xmin = X[:, 0].min()
xmax = X[:, 0].max()
ymin = X[:, 1].min()
ymax = X[:, 1].max()

xx = np.linspace(xmin, xmax, 10)
yy = np.linspace(ymin, ymax, 10)
X1, X2 = np.meshgrid(xx, yy)
Z = np.empty(X1.shape)
for (i, j), val in np.ndenumerate(X1):
    x1 = val
    x2 = X2[i, j]
    p = model.decision_function([[x1, x2]])
    Z[i, j] = p[0]

levels = [-1, 0, 1]
linestyles = ['dashed', 'solid', 'dashed']
plt.scatter(X[y == -1, 0], X[y == -1, 1], marker = 'o', label = '-1 클래스')
plt.scatter(X[y == 1, 0], X[y == 1, 1], marker = 'x', label = '+1 클래스')
plt.contour(X1, X2, Z, levels, colors = 'k', linestyles = linestyles)
plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1], s = 300, alpha = 0.3)

x_new = [10, 2]
plt.scatter(x_new[0], x_new[1], marker = '^', s = 100)
plt.text(x_new[0] + 0.03, x_new[1] + 0.08, '테스트 데이터')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('SVM 예측 결과')
plt.show()
```

```python
np.random.seed(0)
X = np.r_[np.random.randn(20, 2) - [2, 2], np.random.randn(20, 2) + [2 ,2]]
Y = [-1] * 20 + [1] * 20

plotnum = 1
for name, penalty in (('C=10', 10), ('C=0.1', 0.1)):
    clf = SVC(kernel = 'linear', C = penalty).fit(X, Y)
    xx = np.linspace(-5, 5)

    x_jin = -5
    x_jax = 5
    y_jin = -9
    y_jax = 9
    XX, YY = np.mgrid[x_jin:x_jax:200j, y_jin:y_jax:200j]

    levels = [-1, 0, 1]
    linestyles = ['dashed', 'solid', 'dashed']
    Z = clf.decision_function(np.c_[XX.ravel(), YY.ravel()])
    Z = Z.reshape(XX.shape)

    plt.subplot(1, 2, plotnum)
    plt.contour(XX, YY, Z, levels, colors = 'k', linestyles = linestyles)
    plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s = 120, linewidth = 4)
    plt.scatter(X[:, 0], X[:, 1], c = Y, s = 60, linewidth = 1, cmap = plt.cm.Paired)
    plt.xlim(x_jin, x_jax)
    plt.ylim(y_jin, y_jax)
    plt.title(name)

    plotnum += 1

plt.suptitle('슬랙변수 가중치 C의 영향')
plt.tight_layout()
plt.show()
```

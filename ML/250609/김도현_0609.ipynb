# 레이블 인코딩 진행

```python
from sklearn.preprocessing import LabelEncoder

for col in df2.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df2[col] = le.fit_transform(df2[col])

df2
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df2.iloc[:, 1:], df2.iloc[:, 0], test_size = 0.2, random_state = 42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
forest = RandomForestClassifier()
forest.fit(X_train, y_train)
```

#### 모델 정확도

```python
score = forest.score(X_test, y_test)
print('정확도 :', score)
```

#### 훈련 데이터에 대한 정확도

```python
y_train_pred = forest.predict(X_train)
print('훈련 데이터 정확도 :', metrics.accuracy_score(y_train, y_train_pred))
```

#### 테스트 데이터 정확도

```python
y_pred = forest.predict(X_test)
print('정확도 :', metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

# 최적의 하이퍼 파라미터 찾기

```python
from sklearn.model_selection import GridSearchCV, KFold

estimator = RandomForestClassifier()
cv = KFold(n_splits=5, random_state=42, shuffle=True)

parameters = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2'],
    'bootstrap': [True, False],
    'criterion': ['gini', 'entropy']
}

model = GridSearchCV(estimator = estimator,
                    param_grid = parameters,
                    cv = cv, verbose = 1,
                    n_jobs = -1, refit = True) # -1 현재 시스템에서 병렬개수를 최대한 사용하겠다

model.fit(X_train, y_train)
```

```python
import warnings
warnings.filterwarnings(action='ignore')

print("Best Estimator:\n", model.best_estimator_);print()
print("Best Params:\n", model.best_params_);print()
print("Best Score:\n", model.best_score_);print()
```

```python
forest = RandomForestClassifier(
    bootstrap = True,
    criterion = 'gini',
    max_depth = None,
    max_features = 'sqrt',
    min_samples_leaf = 1,
    min_samples_split = 2,
    n_estimators = 100
)

forest.fit(X_train, y_train)
```

#### 훈련 데이터 정확도

```python
y_train_pred = forest.predict(X_train)
print('훈련 데이터 정확도 :', metrics.accuracy_score(y_train, y_train_pred))
```

#### 테스트 데이터 정확도

```python
y_pred = forest.predict(X_test)
print('정확도 :', metrics.accuracy_score(y_test, y_pred))
```

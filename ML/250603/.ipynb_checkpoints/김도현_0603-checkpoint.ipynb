```python
def loss_func(x, t):
    y = np.dot(x, W) + b
    return (np.sum((t - y) ** 2)) / (len(x)) # MSE
```

```python
def numerical_derivative(f, x):
    delta_x = 1e-4
    grad = np.zeros_like(x)
    it = np.nditer(x, flags = ['multi_index'], op_flags = ['readwrite'])
    while not it.finished:
        idx = it.multi_index
        tmp_val = x[idx]
        x[idx] = float(tmp_val) + delta_x
        fx1 = f(x) # f(x + delta_x)

        x[idx] = float(tmp_val) - delta_x
        fx2 = f(x)
        grad[idx] = (fx1 - fx2) / (2 * delta_x)
        x[idx] = tmp_val
        it.iternext()

    return grad
```

```python
def predict(x):
    y = np.dot(x, W) + b
    return y
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

loaded_data = np.loadtxt('../AI데이터/boston_house.csv', delimiter = ',', skiprows = 1, dtype = np.float32)
x_data = loaded_data[:, :-1]
t_data = loaded_data[:, -1]

print('x_data.ndim = ', x_data.ndim, ', x_data.shape = ', x_data.shape)
print('t_data.ndim = ', t_data.ndim, ', t_data.shape = ', t_data.shape)
```

```python
np.random.seed(42)
W = np.random.rand(13, 1)
b = np.random.rand(1)

print('W = ', W, 'W.shape = ', W.shape, ', b = ', b, 'b.shape = ', b.shape)
```

```python
learning_rate = 1e-5
f = lambda x : loss_func(x_data, t_data)
print('Initial loss value = ', loss_func(x_data, t_data))

for step in range(30001):
    W -= learning_rate * numerical_derivative(f, W)
    b -= learning_rate * numerical_derivative(f, b)

    if(step % 1000 == 0):
        print('step = ', step, 'loss value = ', loss_func(x_data, t_data),
              'W = ', W, 'b = ', b)
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd

df = pd.read_csv('../AI데이터/boston_house.csv')
df.head()
```

```python
dfx = df.drop(['Target'], axis = 1)
dfx.head()
```

```python
dfy = df['Target'] # 종속변수
dfy
```

```python
X_train, X_test, y_train, y_test = train_test_split(dfx, dfy, test_size = 0.2, random_state = 42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train) # X_train 데이터의 평균과 표준편차를 계산하여 저장
X_train = scaler.transform(X_train) # 표준 정규 분포 형태로 변환(스케일)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train) # fit 학습 진행
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

# 테스트 데이터의 예측 정확도 평가 낮을수록 좋음
mse_test = mean_squared_error(y_test, pred)
mse_test
```

```python
score = model.score(X_test, y_test) # 회귀모델에서 정확도는 결정계수
print('정확도 : ', score) 
cofficient = model.coef_ # 가중치(계수)
intercept = model.intercept_ # 바이어스
print('계수 : ', cofficient) 
print('절편 : ', intercept)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, pred)
```

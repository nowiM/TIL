```python
import pandas as pd

df = pd.read_csv('./AI데이터/titanic/train.csv')
df
```

# 결측값 처리

```python
age_data = df[['Age']]
age_data
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
df['Age'] = imp_mean.fit_transform(age_data)
```

```python
df['Age'].isnull().mean() * 100
```

```python
embarked_data = df[['Embarked']]
embarked_data
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

imp = SimpleImputer(strategy='most_frequent')
filled = imp.fit_transform(df[['Embarked']])

le = LabelEncoder()
le.fit(filled)
encoded = le.transform(filled)

df['Embarked'] = encoded
```

```python
df['Embarked'].isnull().mean() * 100
```

```python
import pandas as pd

df = pd.read_csv('./AI데이터/titanic/train.csv')
df
```

```python
df['Embarked'].isnull().mean() * 100
```

```python
y = df.Embarked
y
```

```python
le = LabelEncoder()
le.fit(y)
embarked = le.transform(y)
embarked

# 1차원에서 2차원 배열로 변경
embarked_2d = embarked.reshape(-1, 1)
embarked_2d
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

imp_most = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent') # 적절한하게 선택 mean, median, most_frequent 
filled_embarked = imp_most.fit_transform(embarked_2d)
```

```python
df['Embarked'] = filled_embarked.ravel()
```

```python
df['Embarked'].isnull().mean() * 100
```
